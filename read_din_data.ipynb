# 数据读取

```python
import pickle
import pandas as pd
import numpy as np
import random
```

```python
def to_df(file_path):
  with open(file_path, 'r') as fin:
    df = {}
    i = 0
    for line in fin:
      df[i] = eval(line)
      i += 1
      if i <3:
        print(line)
    df = pd.DataFrame.from_dict(df, orient='index')
    return df
```

```python
reviews_df = to_df('D:/workspace/data/elec5.json')
```

```python
with open('D:/workspace/data/reviews.pkl', 'wb') as f:
  pickle.dump(reviews_df, f, pickle.HIGHEST_PROTOCOL)
```

```python
def build_map(df, col_name):
  key = sorted(df[col_name].unique().tolist())
  m = dict(zip(key, range(len(key))))
  print(m)
  df[col_name] = df[col_name].map(lambda x: m[x])
  return m, key
```

# build_dataset

```python
with open('D:/workspace/data/remap.pkl', 'rb') as f:
  reviews_df = pickle.load(f)
  cate_list = pickle.load(f)
  user_count, item_count, cate_count, example_count = pickle.load(f)
```

```python
reviews_group = reviews_df.groupby('reviewerID')
```

```python
train_set = []
test_set = []
for reviewerID, hist in reviews_df.groupby('reviewerID'):  # {user_id:[raw_idx1, raw_idx2....],user_id2:[raw_idx1, raw_idx2]}
  pos_list = hist['asin'].tolist()  # hist中是整个数据表中归属该user的样本的idx数组，将该用户的样本的asin取出，不定长
  def gen_neg():  # 生成负样本序号，只要随机生成的索引不在用户的点击item范围内就行
    neg = pos_list[0]
    while neg in pos_list:
      neg = random.randint(0, item_count-1)
    return neg
  neg_list = [gen_neg() for i in range(len(pos_list))]  # 为该用户的每个点击item都随机生成一个未点击的item索引作为负样本

  for i in range(1, len(pos_list)):
    hist = pos_list[:i]  # 根据该用户的点击item序列生成从1到全长的点击序列
    if i != len(pos_list) - 1:
      train_set.append((reviewerID, hist, pos_list[i], 1))  # i前面的历史点击用于训练，第i个用于预测
      train_set.append((reviewerID, hist, neg_list[i], 0))
    else:
      label = (pos_list[i], neg_list[i])  # 完整序列的最后一个item，只有一个item的直接让如test_set
      test_set.append((reviewerID, hist, label))  # 因为根据兴趣推荐，所以是根据历史点击预测最后一个item
```
