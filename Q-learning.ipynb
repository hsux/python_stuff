```python
import numpy as np
GAMMA = 0.8
Q = np.zeros((6,6))
R=np.asarray([[-1,-1,-1,-1,0,-1],
   [-1,-1,-1,0,-1,100],
   [-1,-1,-1,0,-1,-1],
   [-1,0, 0, -1,0,-1],
   [0,-1,-1,0,-1,100],
   [-1,0,-1,-1,0,100]])
```

```python
alpha = 0.1
```

```python
def getMaxQ(state):
    return max(Q[state, :])
```

```python
def QLearning(state):
    curAction = None
    for action in range(6):
        if(R[state][action] == -1):
            Q[state, action]=0
        else:
            curAction = action
            Q[state,action] = Q[state,action] + alpha*(R[state][action]+GAMMA * getMaxQ(curAction) - Q[state,action])
```

```python
count=0
while count<10:
    for i in range(6):
        QLearning(i)
    count+=1
    print('============{}===={}============='.format(count,i))
    print(Q/5)
    print('===============================')
print(Q/5)
```
